### Data Preprocessing

#### Data Encoding

#### Finding the categorical columns

#### Outlier Detection

```python
df.describe()
```

```python
columns = df.drop(['class'], axis = 1).columns
columns
```

Finding columns with corresponding Z_scores

```python
z_scores = {}
for col in columns:
  z_score = (df[col] - df[col].mean())/df[col].std()
  z_scores[col] = z_score


print(z_scores['air_time1'])
```

Finding indexes for each column where the outliers are found

```python
indexes = {}
for col, data in z_scores.items():
      idx = [i for i,  val in enumerate(data) if val > 3 or val < -3 ]
      indexes[col] = idx

indexes
```

Creating a copy dataframe to store data with updated outliers

```python
df_copy = df.copy()

for col, idx in indexes.items():
  meanVal = df_copy[col].mean()
  for id in idx:
    df_copy[col][id] = meanVal

```

```python
df = df_copy

from sklearn.utils import shuffle
shuffled_df = shuffle(df)
```

```python
rearranged_df = shuffled_df.reset_index(drop=True)
rearranged_df
```

#### Separating Features and Targets

```python
X = rearranged_df.drop(columns = ['class'])
y = rearranged_df['class']
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
min_max_scaler = MinMaxScaler()
min_max_scaled_X = min_max_scaler.fit_transform(X)
min_max_scaled_X
```

```python
STD_scaler = StandardScaler()
STD_scaled_X = STD_scaler.fit_transform(X)
STD_scaled_X
```

### Training the model

#### Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(min_max_scaled_X, y, test_size = 0.20, random_state=4 )
```

```python
print(f"Train Data Size: (Features): {len(x_train)}")
print(f"Train Data Size (Target): {len(y_train)}")
print(f"Test Data Size: (Features): {len(x_test)}")
print(f"Test Data Size: (Target): {len(y_test)}")
```

```python
x_train
```

#### Ensemble Method

#### Defining The models

We will define some classifiers and use them for ensembling

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
model1 = RandomForestClassifier(n_estimators = 10)
model2 = AdaBoostClassifier(n_estimators = 100, algorithm = 'SAMME', random_state = 0)
```

```python
models = {
    "Random Forest": model1,
    "Ada Boost": model2
}
```

#### Making prediction

#### Evaluation Metrics

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

### Comparing with the individual performance of estimators

```python
for name, model in models.items():
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)

  print(f"\n\n\n\n------- Model Name: {name} --------\n\n")
  print("Accuracy: ", accuracy_score(y_pred, y_test))
  print("Precision: ", precision_score(y_pred, y_test, average='macro'))
  print("Recall: ", recall_score(y_pred, y_test, average='macro'))
  print("f1: ", f1_score(y_pred, y_test, average='macro'))
```
